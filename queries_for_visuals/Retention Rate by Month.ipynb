```sql
with retention_table as (
  SELECT  
  customer_id, 
  DATE_FORMAT(DATE_TRUNC('month', first_purchase_date), 'yyyy-MM') AS cohort_month,
  DATE_DIFF(month, first_purchase_date, second_purchase_date) as month_between_purchases,
  -- flag the month between the first and the second purchase
  CASE WHEN month_between_purchases <= 1 THEN 1 ELSE 0 END AS retention_rate_1m,
  CASE WHEN month_between_purchases <= 2 THEN 1 ELSE 0 END as retention_rate_2m,
  CASE WHEN month_between_purchases <= 3 THEN 1 ELSE 0 END as retention_rate_3m
from workspace.bigquery_db_cohort_db.cohort_analysis
group by customer_id, first_purchase_date, second_purchase_date
)

-- retention percentages
select cohort_month, 
       -- calculate the retention rates for each intervall
       round(100 * sum(retention_rate_1m)/count(customer_id),2) as retention_rate_1m,
       round(100 * sum(retention_rate_2m)/count(customer_id),2) as retention_rate_2m,
       round(100 * sum(retention_rate_3m)/count(customer_id),2) as retention_rate_3m
from retention_table
group by cohort_month
ORDER BY cohort_month


 
```
