```sql

with customer_avg as (
    select customer_id,
    date_format(date_trunc("month", min(first_purchase_date)),'yyyy-MM') as cohort_month,
    days_between_purchases
from workspace.bigquery_db_cohort_db.cohort_analysis
group by customer_id, days_between_purchases    
)

select cohort_month, Round(avg(days_between_purchases) / 30.44, 2) as avg_month_between_purchases
from customer_avg
group by cohort_month
```
